```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pylab as pl
import os
pd.set_option("display.width", 3000)
pd.set_option("max_colwidth", -1)
directory=os.path.abspath(os.getcwd())
file=pd.read_csv('result.csv')
# file.drop("Unnamed: 0",axis=1,inplace=True) 
file.index = np.arange(1, len(file)+1)
Open=file['Open']
a=[]
for i in Open:
    a.append(float(i))
a=a[::-1]
a
file
High=file['High']
b=[]
for i in High:
    b.append(float(i))
b=b[::-1]
Low=file['Low']
c=[]
for i in Low:
    c.append(float(i))
c=c[::-1]
Close=file['Close']
d=[]
for i in Close:
    d.append(float(i))
d=d[::-1]
df=pd.DataFrame([a,b,c,d],index=['Open','High','Low','Close'],columns=file['Date'].index)
df
date=[]
date.append(file['Date'].index)
date
for i in date:
    figure=df[i].mean()
pl.figure(figsize=(150, 150))
q=plt.plot(figure)

q=plt.savefig(directory+'/for_father')
```

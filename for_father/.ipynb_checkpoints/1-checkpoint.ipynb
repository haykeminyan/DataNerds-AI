```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
pd.set_option("display.width", 3000)
pd.set_option("max_colwidth", -1)
directory=os.path.abspath(os.getcwd())
file=pd.read_csv('result.csv')
file.drop("Unnamed: 0",axis=1,inplace=True) 
file.index = np.arange(1, len(file)+1)
Open=file['Open']
a=[]
for i in Open:
    a.append(float(i.replace(',','')))

file
High=file['High']
b=[]
for i in High:
    b.append(float(i.replace(',','')))
Low=file['Low']
c=[]
for i in Low:
    c.append(float(i.replace(',','')))
Close=file['Close']
d=[]
for i in Close:
    d.append(float(i.replace(',','')))
df=pd.DataFrame([a,b,c,d],index=['Open','High','Low','Close'],columns=file['Date'])
df
date=[]
date.append(file['Date'])

for i in date:
    figure=df[i].mean()
# plt.figure(figsize=(6*3.13,4*3.13))
q=plt.plot(figure)
# q=plt.figure(points(figsize=(6*3.13,4*3.13)))
q=plt.savefig(directory+'/for_father')
```

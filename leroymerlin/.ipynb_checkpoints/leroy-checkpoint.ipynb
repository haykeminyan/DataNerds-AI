```python
import pymongo
from pymongo import MongoClient
import pandas as pd
import numpy as np
import dateutil.parser
import datetime
pd.set_option("display.width", 3000)
pd.set_option("max_colwidth", -1)

client=MongoClient('localhost',27017)
db = client.shop_leroy
final_2=db.final_2

dictionary={}
data = db.final_2.find({},{'_id':0})
dates=[]

for i in list(data):
    dictionary = {**dictionary, **i}

goods =[] 
href=[]

category=[]
for i in dictionary.keys():
    category.append(i)


subcategory=[]
for i in dictionary.values():
    for j in i.keys():
        subcategory.append(j)


links_and_dates=[]
for i in dictionary.values():
    for j in i.values():
        for k in j.keys():
            links_and_dates.append(k)
links=[]
for i in  links_and_dates:
    if i=='link':
        links.append(i)
    else:
        dates.append(i)
        

for i in dictionary.values():
    for j in i.values():
        for j in j.values():
            if type(j)==dict:
                goods.append(j)
            else:
                href.append(j)

goods_final=[]
for i,j in zip(dates,goods):
    goods_final.append(i+': '+str(j))

final={}



final=[subcategory,goods_final,href]

df = pd.DataFrame.from_dict(final)
df.index=['Категории','Товары','Ссылки']
df.columns+=1
df.T
```

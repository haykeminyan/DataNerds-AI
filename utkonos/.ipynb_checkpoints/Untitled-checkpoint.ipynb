```python
import pymongo
from pymongo import MongoClient
import pandas as pd
import numpy as np
import dateutil.parser
import datetime
pd.set_option("display.width", 3000)
pd.set_option("max_colwidth", -1)

client=MongoClient('localhost',27017)
db=client.test
final=db.final

dictionary={}
data = db.final.find({},{'_id':0})
# print(data)


for i in list(data):
    dictionary = {**dictionary, **i}
# print(dictionary)

goods =[] 
href=[]

category=[]
for i in dictionary.keys():
    category.append(i)
print(category)

subcategory=[]
for i in dictionary.values():
    for j in i.keys():
        subcategory.append(j)
print(subcategory)

links_and_dates=[]
for i in dictionary.values():
    for j in i.values():
        for k in j.keys():
            links_and_dates.append(k)
links=[]
for i in  links_and_dates:
    if i=='link':
        links.append(i)
    else:
        dates.append(i)
        
# print(links)
# print(dates)



for i in dictionary.values():
    for j in i.values():
        for j in j.values():
            if type(j)==dict:
                goods.append(j)
            else:
                href.append(j)
# print(href)
# print(goods)
goods_final=[]
for i,j in zip(dates,goods):
    goods_final.append(i+': '+str(j))
# print(goods_final)
final={}

href_final=[]
for i,j in zip(links,href):
    href_final.append(i+j)
print(href_final)
# final=[category,subcategory,goods,links]

df = pd.DataFrame.from_dict(final)
# df.index = np.arange(1, len(df) + 1)
# df.index=['Категории','Подкатегории','Продукты','Ссылки']
# list(df.index['category','dsadas','ds','s'])
df.T
```

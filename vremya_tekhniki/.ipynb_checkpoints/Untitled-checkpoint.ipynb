```python
import pymongo
from pymongo import MongoClient
import pandas as pd
import numpy as np
import dateutil.parser
import datetime
pd.set_option("display.width", 3000)
pd.set_option("max_colwidth", -1)

client=MongoClient('localhost',27017)
db=client.tekh
final_3=db.final_3

dictionary={}
data = db.final.find({},{'_id':0})
# print(data)
dates=[]

for i in list(data):
    dictionary = {**dictionary, **i}
# print(dictionary)

goods =[] 
href=[]

category=[]
for i in dictionary.keys():
    category.append(i)
# print(category)

subcategory=[]
for i in dictionary.values():
    for j in i.keys():
        subcategory.append(j)
# print(subcategory)

links_and_dates=[]
for i in dictionary.values():
    for j in i.values():
        for k in j.keys():
            links_and_dates.append(k)
links=[]
for i in  links_and_dates:
    if i=='link':
        links.append(i)
    else:
        dates.append(i)
        
# print(links)
# print(dates)



for i in dictionary.values():
    for j in i.values():
        for j in j.values():
            if type(j)==dict:
                goods.append(j)
            else:
                href.append(j)
# print(href)
# print(goods)
goods_final=[]
for i,j in zip(dates,goods):
    goods_final.append(i+': '+str(j))
# print(goods_final)
final={}


# final=[category,subcategory,goods_final,href]
final=[subcategory,goods_final,href]

df = pd.DataFrame.from_dict(final)
# df.index=['Категории','Подкатегории','Продукты','Ссылки']
df.index=['Категории','Продукты','Ссылки']
df.columns+=1
df.T
```

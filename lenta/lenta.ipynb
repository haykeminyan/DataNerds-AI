```python
import pymongo
from pymongo import MongoClient
import pandas as pd
pd.set_option("max_colwidth", 1000)

client = MongoClient('localhost',27017)
db = client.shop_prices
final_lenta = db.final_lenta
dictionary = {}
data = db.final_lenta.find({},{'_id':0})
dates = []
goods = [] 
href = []
category = []
counter = []

for i in list(data):
    category.append(i['category_name'])
    href.append(i['link'])
    dates.append(list(i.keys())[2::2])
    goods.append(list(i.values())[2::2])

for i in range(len(goods)):
    counter.append(len(list((goods[i][0]).keys())))

final = {}
final = [category,goods,counter,href,dates]
df = pd.DataFrame.from_dict(final)
df.index = ['Категории','Товары','Число товаров','Ссылки','Дата']
df.columns += 1
df.T
```

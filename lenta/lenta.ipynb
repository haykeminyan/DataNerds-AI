```python
import pymongo
from pymongo import MongoClient
import pandas as pd

pd.set_option("display.width", 3000)
pd.set_option("max_colwidth", 0)

client=MongoClient('localhost',27017)
db = client.shop_prices
final_lenta=db.final_lenta

dictionary={}
data = db.final_lenta.find({},{'_id':0})

dates = []
goods = [] 
href = []
category = []

for i in list(data):
    category.append(i['category_name'])
    href.append(i['link'])
    dates.append(list(i.keys())[2::2])
    goods.append(list(i.values())[2::2])

final = {}
final = [category,goods,href,dates]
df = pd.DataFrame.from_dict(final)
df.index = ['Категории','Товары','Ссылки','Дата']
df.columns += 1
df.T
```

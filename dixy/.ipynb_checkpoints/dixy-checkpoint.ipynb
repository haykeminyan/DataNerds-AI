```python
import pymongo
from pymongo import MongoClient
import pandas as pd
import numpy as np
import dateutil.parser
import datetime
pd.set_option("display.width", 3000)
pd.set_option("max_colwidth", -1)

client=MongoClient('localhost',27017)
db = client.shop_prices
final_dixy=db.final_dixy
dictionary={}

data = db.final_dixy.find({},{'_id':0})




for i in list(data):
    dictionary = {**dictionary, **i}

goods =[] 
href=[]

category=[]
for i in dictionary.keys():
    category.append(i)


dates=[]
links_and_dates=[]
for i in dictionary.values():
    for k in i.keys():
        links_and_dates.append(k)

links=[]
for i in  links_and_dates:
    if i=='link':
        links.append(i)
    else:
        dates.append(i)


for i in dictionary.values():
    for j in i.values():
        if type(j)==dict:
            goods.append(j)
        else:
            href.append(j)


final={}

final=[category,goods,href]

df = pd.DataFrame.from_dict(final)
df.index=['Категории','Товары на '+dates[0],'Ссылки']
df.columns+=1
df.T
```
